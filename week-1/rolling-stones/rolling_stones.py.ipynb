```python
import csv

albums_dict = []
with open('data.csv') as f:
    reader = csv.DictReader(f)
    albums_dict.extend(list(reader))
albums_dict[:5]
```

```python
def find_by_name(name):
    for album in albums_dict:
        if album['album'] == name:
            return album
    return None
```

```python
print(find_by_name('Rubber Soul'))
print(type(find_by_name('Gibberish')))
```

```python
def find_by_rank(rank):
    if int(rank) < 1 or int(rank) > 500:
        return None
    for album in albums_dict:
        if album['number'] == str(rank):
            return album
    return None
```

```python
print(find_by_rank(5))
print(find_by_rank('5'))
print(type(find_by_rank(-20)))
```

```python
def find_by_year(year):
    retVal = []
    for album in albums_dict:
        if album['year'] == str(year):
            retVal.append(album)
    return retVal
```

```python
print(find_by_year(1965) == find_by_year('1965'))
print(len(find_by_year(2020)))
find_by_year(1965)
```

```python
def find_by_years(s_year,e_year):
    retVal = []
    start = min(int(s_year),int(e_year))
    end = max(int(s_year),int(e_year))
    for album in albums_dict:
        if int(album['year']) >= start and int(album['year']) <= end:
            retVal.append(album)
    return retVal
```

```python
sorted(find_by_years(1990,2000),key=lambda alb:int(alb['year']))
```

```python
def find_by_ranks(s_rank,e_rank):
    retVal = []
    start = min(int(s_rank),int(e_rank))
    end = max(int(s_rank),int(e_rank))
    for album in albums_dict:
        if int(album['number']) >= start and int(album['number']) <= end:
            retVal.append(album)
    return retVal
```

```python
sorted(find_by_ranks(50,61),key=lambda alb:int(alb['number']))
# len(albums_dict)
```

```python
def all_titles():
    retVal = []
    for album in albums_dict:
        retVal.append(album['album'])
    return retVal
```

```python
all_titles()
# len(all_titles())
```

```python
def all_artists():
    retVal = []
    for album in albums_dict:
        retVal.append(album['artist'])
    return retVal
```

```python
all_artists()
# len(all_artists())
```

```python
def most_albums():
    albums_by_artists = {}
    for album in albums_dict:
        if albums_by_artists.get(album['artist']) is None:
            albums_by_artists.update({album['artist']: 1})
        else:
            albums_by_artists[album['artist']] += 1
    top_artists = sorted(albums_by_artists.items(),key=lambda artist: artist[1],reverse=True)
    return top_artists[0][0]
```

```python
most_albums()
```

```python
def most_popular_word():
    word_count = {}
    for album in albums_dict:
        for word in album['album'].split():
            if word_count.get(word) is None:
                word_count.update({word: 1})
            else:
                word_count[word] += 1
    top_words = sorted(word_count.items(),key=lambda words: words[1],reverse=True)
    return top_words[0]
```

```python
most_popular_word()
```
