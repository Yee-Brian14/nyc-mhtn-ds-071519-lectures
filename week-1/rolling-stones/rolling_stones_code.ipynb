```python
import csv

albums_dict = []
with open('data.csv') as f:
    reader = csv.DictReader(f)
    albums_dict.extend(list(reader))
# albums_dict[:5]
```

```python
def find_by_year(year,media_type):
    """
    Returns the albums or songs released in a given year.
    
    Parameters
    ----------
    year : int
        The year to search for
    media_type : str
        Either 'album' or 'song' depending on which one to look for.
    """
    retVal = []
    if media_type.lower() == "album":
        data_set = albums_dict
    elif media_type.lower() == "song":
        data_set = song_list
    else:
        return None
    for media in data_set:
        if media['year'] == str(year):
            retVal.append(media)
    return retVal
```

```python
def find_by_years(s_year,e_year,media_type):
    """
    Returns a list of all albums or songs released during a range of years.
    
    Parameters
    ----------
    s_year : int
        The start year of the range to look through.
    e_year : int
        The end year of the range to look through.
    media_type : str
        Either 'album' or 'song' depending on which one to look for.
    """
    retVal = []
    start = min(int(s_year),int(e_year))
    end = max(int(s_year),int(e_year))
    if media_type.lower() == "album":
        data_set = albums_dict
    elif media_type.lower() == "song":
        data_set = song_list
    else:
        print("Not a valid media type. Enter album or song.")
        return None
    for data in data_set:
        if int(data['year']) >= start and int(data['year']) <= end:
            retVal.append(data)
    return retVal
```

```python
def find_by_ranks(s_rank,e_rank,media_type):
    """
    Returns a list of all albums or songs with ranks in a range.
    
    Parameters
    ----------
    s_rank : int
        The start rank of the range to look through.
    e_rank : int
        The end rank of the range to look through.
    media_type : str
        Either 'album' or 'song' depending on which one to look for.
    """
    retVal = []
    start = min(int(s_rank),int(e_rank))
    end = max(int(s_rank),int(e_rank))
    if media_type.lower() == 'album':
        data_set = albums_dict
        name = 'number'
    elif media_type.lower() == 'song':
        data_set = song_list
        name = 'rank'
    else:
        print("Not a valid media type. Enter album or song.")
        return None
    for data in data_set:
        if int(data[name]) >= start and int(data[name]) <= end:
            retVal.append(data)
    return retVal
```

```python
def all_titles(media_type):
    """
    Returns a list of all album titles or song titles.
    
    Parameters
    ----------
    media_type : str
        Either 'album' or 'song' depending on which one to look for.
    """
    retVal = []
    if media_type.lower() == 'album':
        data_set = album_dict
        name = 'album'
    elif media_type.lower() == 'song':
        data_set = song_list
        name = 'name'
    else:
        print("Not a valid media type. Enter album or song.")
        return None
    for data in data_set:
        retVal.append(data[name])
    return retVal
```

```python
def all_artists(media_type):
    """
    Returns a list of all artists for either albums or songs.
    
    Parameters
    ----------
    media_type : str
        Either 'album' or 'song' depending on which one to look for.
    """
    retVal = []
    if media_type.lower() == 'album':
        data_set = albums_dict
    elif media_type.lower() == 'song':
        data_set = song_list
    else:
        print("Not a valid media type. Enter album or song.")
        return None
    for data in data_set:
#         if not album['artist'] in retVal:
#             retVal.append(album['artist'])
        retVal.append(data['artist'])
    return list(set(retVal))
```

```python
def most_albums_or_songs(media_type):
    """
    Returns the artist with the most albums or songs in the top 500
    
    Parameters
    ----------
    media_type : str
        Either 'album' or 'song'.
    """
    albums_by_artists = dict.fromkeys(all_artists(media_type),0)
    retVal = []
    if media_type.lower() == 'album':
        data_set = albums_dict
    elif media_type.lower() == 'song':
        data_set = song_list
    else:
        print("Not a valid media type. Enter album or song.")
        return None
    for data in data_set:
#         if albums_by_artists.get(album['artist']) is None:
#             albums_by_artists.update({album['artist']: 1})
#         else:
        albums_by_artists[data['artist']] += 1
    top_artists = sorted(albums_by_artists.items(),key=lambda artist: artist[1],reverse=True)
    top_number = top_artists[0][1]
    for artist in top_artists:
        if artist[1] == top_number:
            retVal.append(artist[0])
        elif artist[1] < top_number:
            break
    return retVal
```

```python
def most_popular_word(media_type):
    """
    Returns the most popular word amongst all album or song titles.
    
    Parameters
    ----------
    media_type : str
        Either 'album' or 'song'.
    """
    word_count = {}
    retVal = []
    if media_type.lower() == 'album':
        name = 'album'
        data_set = albums_dict
    elif media_type.lower() == 'song':
        name = 'name'
        data_set = song_list
    else:
        print("Not a valid media type. Enter album or song.")
        return None
    for data in data_set:
        for word in data[name].split():
            if word_count.get(word.lower()) is None:
                word_count.update({word.lower(): 1})
            else:
                word_count[word.lower()] += 1
    top_words = sorted(word_count.items(),key=lambda words: words[1],reverse=True)
    top_number = top_words[0][1]
    for word in top_words:
        if word[1] == top_number:
            retVal.append(word)
        else:
            break
    return retVal
```

```python
def albumWithMostTopSongs():
    """
    Returns the name of the artist and album that has the most songs featured on the top 500 song list.
    
    Parameters
    ----------
    None
    """
    # keeps count of what is currently the most number of songs
    # keeps track of who currently has the record of most number of songs
    # values are lists in case there are multiple albums with the same number of songs
    max_counter = 0
    top_artist_album = []
    # list of top 500 songs
    all_songs = all_titles('song')
    
    # iterate through the albums in json_data
    for album in json_data:
        # for each album, start a new counter to count how many of its songs are in the top 500
        curr_counter = 0
        # iterate through the tracks for the current album
        for track in album['tracks']:
            # if one of the songs in the current album are in the top 500, increment the current counter
            if track in all_songs:
                curr_counter += 1
        # after iterating through the list of tracks on the album, compare the current counter to the max counter
        # if the current counter is greater than the max counter, then that means the current album has more songs
        # in the top 500 than the currently stored album. Thus we update max counter to the higher number and update
        # the top_artist_album to this new album.
        if curr_counter > max_counter:
            max_counter = curr_counter
            top_artist_album = [(album['artist'],album['album'])]
        # if the current counter is equal to the max counter, we will want to add this new album to top_artist_album
        # while still preserving the albums already there
        # if the current counter is lower than the max counter, we will ignore the album
        elif curr_counter == max_counter:
            top_artist_album.append((album['artist'],album['album']))
    # finally return the top album and artist
    return top_artist_album
        
```

```python
def albumsWithTopSongs():
    """
    Returns a list with names of albums with tracks in the Top 500 songs.
    
    Parameters
    ----------
    None
    
    """
    # list to store the album names to be returned
    retVal = []
    # all the top 500 songs
    all_songs = all_titles('song')
    # iterate through each album in json_data
    for album in json_data:
        # iterate through each track of the album
        for track in album['tracks']:
            # if the track is in the top 500, then add the album to the return list and break out of the loop
            if track in all_songs:
                retVal.append(album['album'])
                break
    return retVal
```

```python
def top10AlbumsByTopSongs():
    """
    Returns a histogram with the 10 albums that have the most songs that appear in the
    top songs list. The album names should point to the number of songs that appear on
    the top 500 songs list.
    
    Parameters
    ----------
    None
    
    """
    albums_with_songs = []
    top10albums = []
    all_songs = all_titles('song')
    for album in json_data:
        for track in album['tracks']:
            if track in all_songs:
                albums_with_songs.append(album['album'])
#                 print(album['album'])
    albums_with_songs = sorted(albums_with_songs,key=lambda album: albums_with_songs.count(album),reverse = True)
    num_albums = 0
    for album in albums_with_songs:
        # if the album isn't currently accounted for, add it in if there are less than 10. break if there are
        # already 10.
        if not album in top10albums and num_albums >= 10:
            break
        elif not album in top10albums and num_albums < 10:
            num_albums += 1
            top10albums.append(album)
        else:
            top10albums.append(album)
    
    top10albums = [album[:10]+"..." if len(album)>10 else album for album in top10albums]
    
    plt.figure(figsize=(25,10))
    plt.hist(top10albums,bins = 10,edgecolor = 'black',linewidth = 1.2)
    plt.title("Top 10 Albums by Songs")
    plt.xlabel("Albums")
    plt.ylabel("Number of Songs")
    plt.show()
```

```python
def topOverallArtist():
    """
    Returns the artist with the most albums and songs featured on top 500 lists.
    
    Parameters
    ----------
    None
    """
    every_artist = all_artists('album')
    every_artist.extend(all_artists('song'))
    artist_counter = dict.fromkeys(every_artist,0)
    for album in albums_dict:
        artist_counter[album['artist']] += 1
    for song in song_list:
        artist_counter[song['artist']] += 1
    top_artist = sorted(artist_counter.items(),key=lambda artist: artist[1],reverse=True)
    return top_artist[0][0]
```
